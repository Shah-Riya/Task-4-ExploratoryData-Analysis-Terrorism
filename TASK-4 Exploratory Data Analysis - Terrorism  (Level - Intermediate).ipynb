```python
#Piechart of attack type
plt.figure(figsize=(10,10))
df['attacktype1_txt'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#Bargraph of number of terrorist activities by regions
plt.subplots(figsize=(15,6))
sns.countplot('region_txt',data=df,palette='inferno',order=df['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.title('Terrerist activity by region')
plt.show()
```

```python
#Bargraph of number of terrorist activities by year
plt.subplots(figsize=(15,6))
sns.countplot('iyear',data=df,palette='inferno',order=df['iyear'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.show()
```

```python
#Finding frequency of attacks throughout the year
plt.subplots(figsize=(15,6))
sns.countplot(x=df['iyear'],hue='success',data=df)
plt.xticks(rotation=90)
plt.xlabel('iyear')
plt.show()
```

```python
#Bargraph about most frequently tachnique of attack 
plt.subplots(figsize=(15,6))
sns.countplot('attacktype1_txt',data=df,palette='inferno',order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.show()
```

```python
#Plotting histogram
df.hist(figsize=(35,30))
```

```python
#Double bargraph of countries and attacks
attack=df['country_txt'].value_counts()[:15].to_frame()
attack.columns=['Attacks']
kill=df.groupby('country_txt')['nkillter'].sum().to_frame()
attack.merge(kill,left_index=True,right_index=True).plot.bar(width=0.7)
fig=plt.gcf()
fig.set_size_inches(16,5)
plt.show()
```

```python
#Bargraph of highest terrorism attack in the country
plt.subplots(figsize=(10,20))
sns.barplot(df['country_txt'].value_counts()[:15].index,df['country_txt'].value_counts()[:15].values)
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.show()
```

```python
#Bargraph of highest terrorism attack in the country
plt.subplots(figsize=(50,20))
sns.countplot('country_txt',data=df,palette='inferno',order=df['country_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.show()
```

```python
#Displaying only first five countries
plt.subplots(figsize=(10,10))
years=df.iyear.value_counts().head()
plt.bar(years.index,years.values)
plt.title('Countries')
```

```python
#Displaying only first five attack type
plt.subplots(figsize=(10,10))
attack=df.attacktype1_txt.value_counts().head()
plt.bar(attack.index,years.values)
plt.title('Countries')
```

```python
#Terrorist activities by region in each year by crosstab
pd.crosstab(df.iyear,df.region_txt).plot(kind='area',figsize=(20,10))
plt.title('terrorist acitivities by region in each year')
plt.ylabel('Number of attacks')
plt.show()
```
